# Missing Data

```python
total = df_train.isnull().sum().sort_values(ascending=False)
percent = (df_train.isnull().sum() / len(df_train)).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
missing_data.head(20)
```

```python
print('Before dropping missing values:')
print('Train set:', df_train.shape)
print('Test set:', df_test.shape)
```

```python
df_train = df_train.drop(df_train.loc[df_train['Electrical'].isnull()].index)
df_train.isnull().sum().max()
```

```python
df_test.isnull().sum().max()
```

```python
num_features = df_train.drop(columns='SalePrice').select_dtypes(exclude='object').columns
# num_features
```

```python
cat_features = df_train.select_dtypes(include='object').columns
# cat_features
```

```python
num_imputer = SimpleImputer()
cat_imputer = SimpleImputer(strategy='most_frequent')

num_imputer.fit(df_train[num_features])
cat_imputer.fit(df_train[cat_features])

df_test[num_features] = num_imputer.transform(df_test[num_features]);
df_test[cat_features] = cat_imputer.transform(df_test[cat_features]);
```

```python
df_train = df_train.drop((missing_data[missing_data['Total'] >= 1]).index, 1)
df_test = df_test.drop((missing_data[missing_data['Total'] >= 1]).index, 1)
```

```python
print('After dropping missing values:')
print('Train set:', df_train.shape)
print('Test set:', df_test.shape)
```

```python
assert df_train.isnull().sum().max() == 0
```

```python
assert df_test.isnull().sum().max() == 0
```

# XGBoost

```python
import xgboost as xgb
```

```python
params = {}
params['objective'] = 'reg:squarederror'
params['eta'] = 0.01
params['max_depth'] = 4
```

```python
df_train = pd.concat([df_train, df_y], axis=1)
df_train.shape
```

```python
df_valid = df_train.sample(frac=0.2, random_state=42)
# df_valid
```

```python
df_train = df_train.drop(index=df_valid.index)
# df_train
```

```python
d_train_y = df_train['SalePrice']
d_train_x = df_train.drop('SalePrice', axis=1)

d_valid_y = df_valid['SalePrice']
d_valid_x = df_valid.drop('SalePrice', axis=1)
```

```python
d_train = xgb.DMatrix(d_train_x, label=d_train_y)
d_valid = xgb.DMatrix(d_valid_x, label=d_valid_y)
d_test = xgb.DMatrix(df_test)
```

```python
d_train.num_col(), d_valid.num_col(), d_test.num_col()
```

```python
watchlist = [(d_train, 'train'), (d_valid, 'valid')]
```

```python
bst = xgb.train(params, d_train, 800, watchlist, early_stopping_rounds=50, verbose_eval=40)
```

```python
test_pred = bst.predict(d_test)
test_pred
```

```python
xgb_df = pd.DataFrame({'Id': test_id, 'SalePrice': np.exp(test_pred)})
xgb_df
```

```python
xgb_df.to_csv('output/xgb_submission.csv', index=False)
```
